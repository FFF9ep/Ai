```python
from nltk.stem.porter import PorterStemmer
from nltk.stem.lancaster import LancasterStemmer
from nltk.stem.snowball import SnowballStemmer
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
input_words = ['writing','writen','wrote','friends','randomnize']
```

```python
porterStemmer_obj = PorterStemmer()
lancasterStemmer_obj = LancasterStemmer()
snowballStemmer_obj = SnowballStemmer('english')

wordNetLemmatizer_obj = WordNetLemmatizer()
```

```python
import pandas as pd
import nltk
nltk.download('wordnet')
```

```python
porter_result = [ porterStemmer_obj.stem(w) for w in input_words ]
lancaster_result = [ lancasterStemmer_obj.stem(w) for w in input_words ]
snowball_result = [ snowballStemmer_obj.stem(w) for w in input_words ]

lemmatize_result_v = [ wordNetLemmatizer_obj.lemmatize(w, pos='v') for w in input_words ]
lemmatize_result_n = [ wordNetLemmatizer_obj.lemmatize(w, pos='n') for w in input_words ]
```

```python
pd.DataFrame({
    'input_word' : input_words,
    'porter' : porter_result,
    'lancaster' : lancaster_result,
    'snowball' : snowball_result,
    'lemmatizer_verb' : lemmatize_result_v,
    'lemmatizer_noun' : lemmatize_result_n
})
```

Stemming :
- only Rules (Prefix, Suffix, Infix), less accurate
- fast speed

Lemmatize :
- Check Database, more accurate
- slow speed
